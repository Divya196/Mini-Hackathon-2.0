```python
#Aggregating vechicle class based on vechicle type and weight type.
#Function for labelling
vehicle_class =  pd.Series(ev_registration_by_category_df['Vehicle Class'])

def vehicle_type(vehicle_class):
    if "MOTOR" in vehicle_class:
        return "MOTOR"
    elif "PASSENGER" in vehicle_class:
        return "PASSENGER"
    elif "INVALID CARRIAGE" in vehicle_class:
        return "INVALID CARRIAGE"
    elif "TWO WHEELER" in vehicle_class:
        return "TWO WHEELER"
    elif "THREE WHEELER" in vehicle_class:
        return "THREE WHEELER"
    else:
        return "Other"

def weight_type(vehicle_class):
    if "HEAVY" in vehicle_class:
        return "HEAVY"
    elif "MEDIUM" in vehicle_class:
        return "MEDIUM"
    elif "LIGHT" in vehicle_class:
        return "LIGHT"
    else:
        return "Other"
```

```python
vehicle_class
```

```python
#Creating two new column
#type(vehicle_class)
ev_registration_by_category_df['Vehicle Type'] = ev_registration_by_category_df['Vehicle Class'].apply(vehicle_type)
ev_registration_by_category_df['Weight Type'] = ev_registration_by_category_df['Vehicle Class'].apply(weight_type)
```

```python
#Total registration valye based on vehicle type
registration_sum_vehicletype = ev_registration_by_category_df.groupby('Vehicle Type')['Total Registration'].sum()
registration_sum_vehicletype.sort_values(ascending= False)
```

```python

registration_sum_weighttype = ev_registration_by_category_df.groupby('Weight Type')['Total Registration'].sum()
registration_sum_weighttype.sort_values(ascending= False)
```
